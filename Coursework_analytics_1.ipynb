Для того, чтобы проверить, что вы правильно соединили все датасеты в основной (без `students.csv`), запустите аналогичный приведенному ниже запрос. Выдача должна совпадать. Имейте в виду, что названия некоторых столбцов были переименованы.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime
import matplotlib
```

Создаем датасеты по имеющимся csv файлам. Объединим два файла с использованием 'outer'
   чтобы не потерять информацию. В получившимся датасете мы видим пустые непересекающиеся
   значения в количестве 5 шт. На результаты данное количествот не окажет заметного влияния (всего
   строк 350 682). Поэтому сделаем объединение при помощи 'inner'. Аналогично проведем объединение
   остальных датасетов.

```python


data_progress_phases = pd.read_csv('progress_phases.csv')
data_courses = pd.read_csv('courses.csv')
data_students = pd.read_csv('students.csv')
data_course_contents = pd.read_csv('course_contents.csv')
data_progresses = pd.read_csv('progresses.csv')

progresses_progress_phases = pd.merge(left=data_progress_phases, right=data_progresses,
                                      left_on='progress_id', right_on='id',how='left')
#print(progresses_progress_phases.isnull().sum())
#(progresses_progress_phases.shape)
data_course_contents
```

```python
new_courses = pd.merge(left=progresses_progress_phases, right=data_courses,
                                      left_on='course_id', right_on='id',how='outer')

new_courses.info()
```

```python
# Уберем лишние столбцы: id_x, id_y (дублирующие ) и Unnamed (не информативный)
new_courses = new_courses.drop(columns=['id_x', 'id_y', 'Unnamed: 0'])
new_courses.info()
```

```python
# В данном случае целесообразно объединение проводить по 3-м ключам,т.к имеются 3 совпадающих поля
# иначе получается жуткий датасет весом 16 гигабайт.(попробовал:)) )
data = pd.merge(new_courses, data_course_contents,
                on=['course_id', 'module_number', 'lesson_number'])
# полученный датасет запишем в отдельный файл, чтобы при последующих запусках не дублировать весь
# процесс объединений
data.to_csv('Data.csv')
data.info()
```

```python
 Рассчитать минимальный, максимальный, средний, медианный возраст студентов
Проверим информацию о содержании файла students.csv с помощью data_students.info(). 
Из 43830 зарегистрированных студентов только  25499 указали дату рождения.
При парсинге выходит ошибка выхода за пределы диапазона, используем "coerce". Корректную дату
указали 25490 студентов.Избавляемся от некорректных значений и  добавляем поле возраст.
```

```python
data_students['birthday'] = pd.to_datetime(data_students['birthday'], errors='coerce')
data_students.dropna(subset=['birthday'], inplace=True)
data_students['age'] = 2021 - data_students['birthday'].dt.year
```

```python
При анализе поля age (describe()) видно, что 155 студентов указали возраст неправильно
(очень почтенный возраст либо еще неродившиеся). Удаляем эти значения и на выходе получаем:
Минимальный и максимальный возраст по установленному нами условию 15 и 65 лет соответственно.
средний и медианный возраст 30,6 и 30 лет соответственно.
```

```python
data_students = data_students[(data_students['age'] >14) & (data_students['age'] < 66)]
#data_students.info()
s_min, s_max, s_median, s_mean = data_students["age"].describe()[["min", "max", "50%", "mean"]]
print('min= {0}\nmax= {1}\nmedian= {2}\nmean= {3}'.format(s_min, s_max, s_median, s_mean))
```

```python
# Рассчитать минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
'''
   Объединяем два датасета. Группируем по курсам и выводим результаты min, max, median и mean 
'''
data_data_students = pd.merge(left=data, right=data_students, left_on='student_id', right_on='id')
data_data_students.groupby('title')['age'].describe()[["min", "max", "50%", "mean"]]
```

# 4.1 **(p)**
Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).




```python
# Возьмем файл, с вычисленными номерами   первой и последней дз по плану модуля из п. 3.1
student_summ = pd.read_csv('min_max_course')

# Подсчитаем количество студентов по модулям,начавших ДЗ, отсортируем по номеру модуля и по курсам
student_start = student_summ.groupby(['title', 'module_title', 'module_number', 'field']). agg({'student_id': 'count'})
student_start = pd.DataFrame(student_start).reset_index()
student_start = student_start.sort_values(by=['title', 'module_number'])
student_start.rename(columns={'student_id': 'student_start'}, inplace=True)

# Подсчитаем количество студентов по модулям,закончивших ДЗ, отсортируем по номеру модуля и по курсам
student_summ = student_summ[student_summ['status'] == 'done']
student_finish = student_summ.groupby(['title', 'module_title', 'module_number']). agg({'student_id': 'count'})
student_finish = pd.DataFrame(student_finish).reset_index()
student_finish = student_finish.sort_values(by=['title', 'module_number'])
student_finish.rename(columns={'student_id': 'student_finish'}, inplace=True)

# Объединяем DF и рассчитываем конверсию 
conversion = pd.merge(student_start, student_finish, on=['title', 'module_title', 'module_number' ])

def mf (module_number):
    if module_number == 1:        
        return  0
    else:
        return 1
conversion['first'] = conversion['module_number'].apply(lambda module_number: mf(module_number))
conversion['student_start'] = conversion['student_start'] * conversion['first']
conversion.student_start = conversion.student_start.shift(-1)
conversion['conversion'] = conversion['student_start'] / conversion['student_finish']
conversion.conversion = conversion.conversion.shift(1)
conversion["conversion"].fillna(0, inplace = True)

conversion.head(12)

```

# 4.2(m)
Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы

```python
conversion

plt.figure(figsize=(80, 6))

plt.bar(conversion['module_title'].values, conversion['conversion'].values)
plt.xlabel('Название модулей')
plt.ylabel('Конверсия')
plt.title('Конверсия перехода студентов из одгного модуля в другой по всем курсам')
plt.yticks()
plt.xticks(size=8, rotation=30, ha='right')
plt.axhline(y=1, ls='--', color='g')

plt.show()
```

# 4.3(m)
Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐

```python
# Создадим объединенный столбец 'module_title'+'module_number'
conversion['m_title_number'] = conversion['module_title'] + ' ' + conversion['module_number'].astype(str)

# Рассчитаем медиану коверсии для каждого курса
median_all = conversion.groupby('title')['conversion'].describe()['50%'].to_dict()
title = conversion['title'].unique().tolist()

# Присвоим сферам цвета 
field_all = conversion.groupby('title')['field'].unique().apply(list).to_dict()

def color_set(field):
    if field == ['Marketing']:
        return 'c'
    elif field == ['Development']:
        return 'g'
    elif field == ['Design']:
        return 'b'
    else:
        return 'r'
fig, axs = plt.subplots(nrows=15, figsize=(5,100))  
   
for i in range(len(title)):    
    median = median_all[title[i]]
    field = field_all[title[i]]
    colors = color_set(field)    
    c = conversion[conversion['title'] == title[i]]
    
    x = c['m_title_number'].values 
    y = c['conversion'].values

    ax = axs[i]
    ax.barh(x, y, alpha=0.1, color=colors)
    ax.set_title(title[i], size=25)
    ax.axvline(x=median, ls='--', color='blue', alpha=0.5, label='median {}'.format(median))
    
    # Удаляем рамки
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    
    # Добавляем легенду
    props_1 = dict(boxstyle='round', facecolor='cyan', alpha=0.1)
    ax.text(1.5, 4, 'marketing', bbox=props_1, color='black', fontsize=10)
    props_2 = dict(boxstyle='round', facecolor='green', alpha=0.1)
    ax.text(1.5, 3, 'Development', bbox=props_2, color='black', fontsize=10)
    props_3 = dict(boxstyle='round', facecolor='blue', alpha=0.1)
    ax.text(1.5, 2, 'design', bbox=props_3, color='black', fontsize=10)
    props_4 = dict(boxstyle='round', facecolor='red', alpha=0.1)
    ax.text(1.5, 1, 'business', bbox=props_4, color='black', fontsize=10)
    ax.text(1.5, 5, 'Сфера курсов:', color='black', fontsize=13)
    ax.legend(loc='upper right')
       
plt.show()


```
